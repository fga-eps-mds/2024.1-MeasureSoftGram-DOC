# Dashboard: Gantt Planejado x Realizado (DA-R1)

Este notebook contém a implementação do gráfico de Gantt para o critério R1 II, comparando os marcos planejados e realizados das Sprints 1 a 5.

## 1. Configuração e Dados

Utilizamos as datas definidas no documento de liderança. 

**Aviso:** Devido ao uso inicial do Zenhub, não foi possível extrair métricas de execução reais automatizadas. Para o gráfico de Gantt da R1, as datas realizadas foram espelhadas das planejadas.

```python
import pandas as pd
import plotly.express as px
import os

# Dados das sprints
data = [
    dict(Sprint='Sprint 1', Start='2026-03-23', Finish='2026-03-30', Tipo='Planejado'),
    dict(Sprint='Sprint 1', Start='2026-03-23', Finish='2026-03-30', Tipo='Realizado'),
    dict(Sprint='Sprint 2', Start='2026-03-30', Finish='2026-04-06', Tipo='Planejado'),
    dict(Sprint='Sprint 2', Start='2026-03-30', Finish='2026-04-06', Tipo='Realizado'),
    dict(Sprint='Sprint 3', Start='2026-04-06', Finish='2026-04-13', Tipo='Planejado'),
    dict(Sprint='Sprint 3', Start='2026-04-06', Finish='2026-04-13', Tipo='Realizado'),
    dict(Sprint='Sprint 4', Start='2026-04-13', Finish='2026-04-20', Tipo='Planejado'),
    dict(Sprint='Sprint 4', Start='2026-04-13', Finish='2026-04-20', Tipo='Realizado'),
    dict(Sprint='Sprint 5', Start='2026-04-20', Finish='2026-04-27', Tipo='Planejado'),
    dict(Sprint='Sprint 5', Start='2026-04-20', Finish='2026-04-27', Tipo='Realizado')
]

df = pd.DataFrame(data)
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

display(df)
```

## 2. Visualização Gantt

Criação do gráfico de linha do tempo.

```python
fig = px.timeline(df, x_start='Start', x_end='Finish', y='Sprint', color='Tipo',
                title='Gantt: Sprints Planejadas vs. Realizadas (DA-R1)',
                labels={'Tipo': 'Status', 'Sprint': 'Sprint'},
                barmode='group')

fig.update_yaxes(autorange='reversed')
fig.update_layout(xaxis_title='Data', yaxis_title='Sprints')
fig.show()
```
